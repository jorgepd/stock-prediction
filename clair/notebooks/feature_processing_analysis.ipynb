```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
np.random.seed(456)
aux = np.arange(50) + np.random.randn(50)*0.2
series1 = np.log(aux)
series2 = np.log(aux[-1] + np.random.randn(40)*0.2)
series = np.append(series1, series2)
series = series[10:]
df = pd.DataFrame(series, columns=['densidade'])

plt.ylabel('densidade')
plt.yticks([])
plt.xlabel('tempo')
plt.xticks([])
plt.plot(df['densidade'])
plt.savefig('price_density.png')
plt.show()
```

```python
series1 = np.arange(50) + np.random.randn(50)*0.2
series2 = series1[-1] + np.random.randn(40)*0.2
series = np.append(series1, series2)
series = series[10:]
df['feature'] = series

plt.ylabel('feature')
plt.yticks([])
plt.xlabel('tempo')
plt.xticks([])
plt.plot(df['feature'])
plt.savefig('feature_original.png')
plt.show()
```

```python
df['time_diff'] = (df['feature'] - df['feature'].shift(1))
df = df[1:]

plt.ylabel('time diff')
# plt.yticks([])
plt.xlabel('tempo')
plt.xticks([])
plt.plot(df['time_diff'])
plt.savefig('feature_time_diff.png')
plt.show()
```

```python
df['ratio'] = (df['time_diff'] / df['feature'])

plt.ylabel('ratio')
plt.yticks([])
plt.xlabel('tempo')
plt.xticks([])
plt.plot(df['ratio'])
plt.savefig('feature_ratio.png')
plt.show()
```
